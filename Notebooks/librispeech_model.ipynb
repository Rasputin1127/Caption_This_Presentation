```python
import os
import librosa   #for audio processing
import IPython.display as ipd
import matplotlib.pyplot as plt
import numpy as np
from scipy.io import wavfile #for audio processing
import warnings
warnings.filterwarnings("ignore")
```

```python
train_audio_path = 'data/LibriSpeech/dev-clean/1272/128104/'
samples, sample_rate = librosa.load(train_audio_path+'1272-128104-0000.wav', sr = 16000)
fig = plt.figure(figsize=(14, 8))
ax1 = fig.add_subplot(211)
ax1.set_title('Raw wave of ' + '1272-128104-0000.wav')
ax1.set_xlabel('time')
ax1.set_ylabel('Amplitude')
ax1.plot(np.linspace(0, len(samples)/sample_rate, len(samples)), samples)
```

```python
resample = librosa.resample(samples, 16000, 8000, fix=True)
```
